```python
categorical_columns = ['loan_type','loan_purpose','Gender','loan_limit','business_or_commercial','co-applicant_credit_type', 
                       'occupancy_type','Region']
for col in categorical_columns:
    df[col] = df[col].astype('category')
```

```python
df['year'] = pd.to_datetime(df['year'], format='%Y')
```

```python
numerical_columns = ['loan_amount', 'rate_of_interest', 'Upfront_charges', 
                     'income', 'property_value', 'LTV']
for col in numerical_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Fill missing values in categorical columns with mode or unknown
df['loan_limit'].fillna(df['loan_limit'].mode()[0], inplace=True)
df['loan_purpose'].fillna(df['loan_purpose'].mode()[0], inplace=True)
```

```python
# Mean/Median Imputation for numerical columns
df['rate_of_interest'].fillna(df['rate_of_interest'].mean(),inplace=True)
df['Upfront_charges'].fillna(df['Upfront_charges'].median(),inplace=True)
df['property_value'].fillna(df['property_value'].median(),inplace=True)
df['income'].fillna(df['income'].median(),inplace=True)
df['LTV'].fillna(df['LTV'].median(),inplace=True)
```

```python
# Choose columns
category_col = 'Region'
value_col = 'loan_amount'

# Group by the categorical column
grouped = df.groupby(category_col)

# Iterate through groups
for category, group in grouped:
    print(f"\nCategory: {category}")

    # Calculate Q1, Q3, and IQR
    Q1 = group[value_col].quantile(0.25)
    Q3 = group[value_col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

    # Identify outliers
    outliers = group[(group[value_col] < lower_bound) | (group[value_col] > upper_bound)]
    print(f"Outliers in {category}:\n{outliers[[value_col]]}")
```

```python
# Split data by Status
defaulted = df[df['Status'] == 1]['loan_amount']
not_defaulted = df[df['Status'] == 0]['loan_amount']

# Perform T-test
t_stat, p_value = stats.ttest_ind(defaulted, not_defaulted)

# Results
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in loan amount between defaulted and non-defaulted loans.")
else:
    print("Fail to reject the null hypothesis: No significant difference in loan amount between defaulted and non-defaulted loans.")
```

```python
# Split data by Status
defaulted_income = df[df['Status'] == 1]['income']
not_defaulted_income = df[df['Status'] == 0]['income']

# Perform T-test
t_stat, p_value = stats.ttest_ind(defaulted_income.dropna(), not_defaulted_income.dropna())

# Results
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in income between defaulted and non-defaulted loans.")
else:
    print("Fail to reject the null hypothesis: No significant difference in income between defaulted and non-defaulted loans.")
```

```python

# Split data by Status
defaulted_score = df[df['Status'] == 1]['Credit_Score']
not_defaulted_score = df[df['Status'] == 0]['Credit_Score']

# Perform T-test
t_stat, p_value = stats.ttest_ind(defaulted_score, not_defaulted_score, nan_policy='omit')

# Results
print(f"T-statistic: {t_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in credit score between defaulted and non-defaulted loans.")
else:
    print("Fail to reject the null hypothesis: No significant difference in credit score between defaulted and non-defaulted loans.")
```

```python

# Create contingency table
contingency_table = pd.crosstab(df['loan_type'], df['Status'])

# Perform Chi-square test
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Results
print(f"Chi2 Statistic: {chi2}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant association between loan type and loan status.")
else:
    print("Fail to reject the null hypothesis: No significant association between loan type and loan status.")
```

```python
from scipy.stats import chi2_contingency

# Contingency table
contingency_table = pd.crosstab(df['loan_purpose'], df['Status'])

# Perform Chi-Square Test
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Results
print(f"Chi2 Statistic: {chi2}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant association between loan purpose and loan default status.")
else:
    print("Fail to reject the null hypothesis: No significant association between loan purpose and loan default status.")
```

```python

# Drop rows with missing values in the columns of interest
filtered_data = df[['loan_amount', 'property_value']].dropna()

# Extract cleaned columns
loan_amount_clean = filtered_data['loan_amount']
property_value_clean = filtered_data['property_value']

# Perform Pearson's correlation
corr, p_value = stats.pearsonr(loan_amount_clean, property_value_clean)

# Results
print(f"Pearson Correlation Coefficient: {corr}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant correlation between loan amount and property value.")
else:
    print("Fail to reject the null hypothesis: No significant correlation between loan amount and property value.")
```

```python

# Split data by Status
defaulted_interest = df[df['Status'] == 1]['rate_of_interest']
not_defaulted_interest = df[df['Status'] == 0]['rate_of_interest']

# Perform Mann-Whitney U Test
u_stat, p_value = stats.mannwhitneyu(defaulted_interest, not_defaulted_interest, alternative='two-sided')

# Results
print(f"U-statistic: {u_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in rate of interest between defaulted and non-defaulted loans.")
else:
    print("Fail to reject the null hypothesis: No significant difference in rate of interest between defaulted and non-defaulted loans.")
```

```python

# Split data by loan_purpose
groups = [group['loan_amount'].dropna() for name, group in df.groupby('loan_purpose')]

# Perform ANOVA
f_stat, p_value = stats.f_oneway(*groups)

# Results
print(f"F-statistic: {f_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in loan amounts across loan purposes.")
else:
    print("Fail to reject the null hypothesis: No significant difference in loan amounts across loan purposes.")
```

```python

# Split data by loan_purpose
groups = [group['rate_of_interest'].dropna() for name, group in df.groupby('loan_purpose')]

# Perform ANOVA
f_stat, p_value = stats.f_oneway(*groups)

# Results
print(f"F-statistic: {f_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in rate of interest across loan purposes.")
else:
    print("Fail to reject the null hypothesis: No significant difference in rate of interest across loan purposes.")
```

```python

# Split data by Status
defaulted_charges = df[df['Status'] == 1]['Upfront_charges']
not_defaulted_charges = df[df['Status'] == 0]['Upfront_charges']

# Perform Mann-Whitney U Test
u_stat, p_value = stats.mannwhitneyu(defaulted_charges.dropna(), not_defaulted_charges.dropna())

# Results
print(f"U-statistic: {u_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference in upfront charges between defaulted and non-defaulted loans.")
else:
    print("Fail to reject the null hypothesis: No significant difference in upfront charges between defaulted and non-defaulted loans.")
```

```python
# Filter out null values for rate_of_interest and loan_amount
data_filtered = df.dropna(subset=['rate_of_interest', 'loan_amount', 'Region'])

# Group by Region for rate_of_interest
groups_interest = [group['rate_of_interest'] for _, group in data_filtered.groupby('Region')]

# Perform ANOVA for rate_of_interest
f_stat_interest, p_value_interest = stats.f_oneway(*groups_interest)
print(f"Rate of Interest - F-statistic: {f_stat_interest}, P-value: {p_value_interest}")
if p_value_interest < 0.05:
    print("Reject the null hypothesis: Significant difference in rate of interest across regions.")
else:
    print("Fail to reject the null hypothesis: No significant difference in rate of interest across regions.")

# Group by Region for loan_amount
groups_amount = [group['loan_amount'] for _, group in data_filtered.groupby('Region')]

# Perform ANOVA for loan_amount
f_stat_amount, p_value_amount = stats.f_oneway(*groups_amount)
print(f"Loan Amount - F-statistic: {f_stat_amount}, P-value: {p_value_amount}")
if p_value_amount < 0.05:
    print("Reject the null hypothesis: Significant difference in loan amounts across regions.")
else:
    print("Fail to reject the null hypothesis: No significant difference in loan amounts across regions.")
```
