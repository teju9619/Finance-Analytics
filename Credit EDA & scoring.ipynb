```python
df['Num_of_Loan'] = df['Num_of_Loan'].astype(str)

# Replace NaN values with an empty string 
df['Num_of_Loan'] = df['Num_of_Loan'].replace('nan', '')  
df['Num_of_Loan'] = df['Num_of_Loan'].replace('', '0')   

# Clean the Num_of_Loan column by removing non-numeric characters
df['Num_of_Loan'] = df['Num_of_Loan'].str.replace(r'\D', '', regex=True)  

# Convert to integer
df['Num_of_Loan'] = df['Num_of_Loan'].astype(int)
```

```python
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].astype(str)

# Replace NaN values with an empty string
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].replace('nan', '') 
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].replace('', '0') 

# Clean the Num_of_Delayed_Payment column by removing non-numeric characters
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].str.replace(r'\D', '', regex=True)  

# Fill NaN values with 0 before converting to integers
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].replace('', '0')  

# Convert to integer
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].astype(int)
```

```python
df['DTI'] = (df['Outstanding_Debt'] + df['Total_EMI_per_month']) / df['Annual_Income'] * 100
df['Savings_Ratio'] = df['Amount_invested_monthly'] / df['Monthly_Inhand_Salary'] * 100
df['Total_Open_Accounts'] = df['Num_Bank_Accounts'] + df['Num_Credit_Card'] + df['Num_of_Loan']
df['Average_Delay'] = df['Delay_from_due_date'] / (df['Num_of_Delayed_Payment'].replace(0, 1))  
```

```python
def convert_age_to_months(age_str):
    if pd.isnull(age_str) or age_str == '':
        return 0  # Handle NaN or empty strings
    years = 0
    months = 0
    parts = age_str.split()
    for i in range(len(parts)):
        if parts[i] == 'Years':
            years = int(parts[i - 1])  
        elif parts[i] == 'Months':
            months = int(parts[i - 1])  
    total_months = years * 12 + months
    return total_months

# Apply the conversion function to the column
df['Credit_History_Age'] = df['Credit_History_Age'].apply(convert_age_to_months)

# Convert to integer
df['Credit_History_Age'] = df['Credit_History_Age'].astype(int)
```

```python
aggregated_df = df.groupby('Customer_ID', as_index=False).agg({
    'Annual_Income': 'mean',
    'Monthly_Inhand_Salary': 'mean',
    'Outstanding_Debt': 'sum',
    'DTI': 'mean',
    'Total_EMI_per_month': 'mean',
    'Amount_invested_monthly': 'mean',
    'Num_Bank_Accounts': 'mean',
    'Num_Credit_Card': 'mean',
    'Num_of_Loan': 'mean',
    'Delay_from_due_date': 'mean',
    'Num_of_Delayed_Payment': 'mean',
    'Credit_History_Age': 'mean',
    'Savings_Ratio': 'mean',
    'Total_Open_Accounts': 'sum',
    'Average_Delay': 'mean',
    'Credit_Utilization_Ratio': 'mean',
    'Payment_of_Min_Amount': 'first'})
```

```python
# feature importance based on hypothetical weighting
weights = {
    'DTI': 0.30,                   # 30% weight for Debt-to-Income ratio (Credit Utilization)
    'Savings_Ratio': 0.10,          # 10% weight for Savings Ratio
    'Total_Open_Accounts': 0.10,    # 10% weight for Total Open Accounts
    'Average_Delay': 0.35,          # 35% weight for Payment History (Delays)
    'Credit_History_Age': 0.15      # 15% weight for Credit Age
}

# Normalize DTI (Lower is better)
aggregated_df['DTI_normalized'] = 1 - (aggregated_df['DTI'] / aggregated_df['DTI'].max())

# Normalize Savings Ratio (Higher is better)
aggregated_df['Savings_Ratio_normalized'] = aggregated_df['Savings_Ratio'] / aggregated_df['Savings_Ratio'].max()

# Normalize Total Open Accounts 
median_accounts = aggregated_df['Total_Open_Accounts'].median()
aggregated_df['Total_Open_Accounts_normalized'] = 1 - np.abs(aggregated_df['Total_Open_Accounts'] - median_accounts) / median_accounts

# Normalize Average Delay
aggregated_df['Average_Delay_normalized'] = 1 - (aggregated_df['Average_Delay'] / aggregated_df['Average_Delay'].max())

# Normalize Credit History Age 
aggregated_df['Credit_History_Age_normalized'] = aggregated_df['Credit_History_Age'] / aggregated_df['Credit_History_Age'].max()

# Now calculate the weighted score based on these normalized values and the assigned weights
aggregated_df['Credit_Score'] = (
    (aggregated_df['DTI_normalized'] * weights['DTI']) +
    (aggregated_df['Savings_Ratio_normalized'] * weights['Savings_Ratio']) +
    (aggregated_df['Total_Open_Accounts_normalized'] * weights['Total_Open_Accounts']) +
    (aggregated_df['Average_Delay_normalized'] * weights['Average_Delay']) +
    (aggregated_df['Credit_History_Age_normalized'] * weights['Credit_History_Age'])
) * 850  # Final score scaled to be out of 850

aggregated_df['Credit_Score'] = aggregated_df['Credit_Score'].round(0)

final_score_df = aggregated_df[['Customer_ID', 'Credit_Score']]

print(final_score_df)
```

```python
# Filter data for last 3 and 6 months
# Assuming df['Month'] is in datetime format and df['Customer_ID'] is the identifier
last_3m = df[df['Month'] >= (df['Month'].max() - pd.DateOffset(months=3))]
last_6m = df[df['Month'] >= (df['Month'].max() - pd.DateOffset(months=6))]

# Aggregating features over last 3 and 6 months
aggregated_3m = last_3m.groupby('Customer_ID').agg({
    'Outstanding_Debt': 'mean',
    'Credit_Utilization_Ratio': 'mean',
    'Num_of_Delayed_Payment': 'sum',
    'Monthly_Inhand_Salary': 'mean',
    'Total_EMI_per_month': 'mean',
    'Num_Credit_Inquiries': 'sum'
})

aggregated_6m = last_6m.groupby('Customer_ID').agg({
    'Outstanding_Debt': 'mean',
    'Credit_Utilization_Ratio': 'mean',
    'Num_of_Delayed_Payment': 'sum',
    'Monthly_Inhand_Salary': 'mean',
    'Total_EMI_per_month': 'mean',
    'Num_Credit_Inquiries': 'sum'
})

# Example: Weighted credit score using recent behavior
credit_score_3m = (aggregated_3m['Outstanding_Debt_normalized'] * 0.3 +
                   aggregated_3m['Credit_Utilization_Ratio_normalized'] * 0.3 +
                   aggregated_3m['Num_of_Delayed_Payment_normalized'] * 0.2 +
                   aggregated_3m['Num_Credit_Inquiries_normalized'] * 0.2)

credit_score_6m = (aggregated_6m['Outstanding_Debt_normalized'] * 0.2 +
                   aggregated_6m['Credit_Utilization_Ratio_normalized'] * 0.2 +
                   aggregated_6m['Num_of_Delayed_Payment_normalized'] * 0.3 +
                   aggregated_6m['Num_Credit_Inquiries_normalized'] * 0.3)

# Final score (you can average the scores or apply a higher weight to recent behavior)
final_credit_score = (credit_score_3m * 0.6) + (credit_score_6m * 0.4)
```
